# Preparação das variáveis

```python
import os
import pymysql
import pandas as pd

host = 'localhost'
password = 'changes'
database = 'trace'

conn = pymysql.connect(
    host=host,
    port=int(3306),
    user="root",
    passwd=password,
    db=database,
    charset='utf8mb4')

work_time = pd.read_sql_query("select TIME_TO_SEC(TIMEDIFF(time_stop,time_start)) worktime FROM perfil where id = 1;",conn).iloc(0)

hi_worktime = int(work_time[0])
md_worktime = int(int(work_time[0]) * 0.50)
lw_worktime = int(int(work_time[0]) * 0.10)

hi_processtime = int(hi_worktime * 0.80)
md_processtime = int(hi_processtime * 0.50)
lw_processtime = int(hi_processtime * 0.20)


process_time_detail = pd.read_sql_query("select process_name, max(cpu_times) as cpu_time from trace_process_detail where process_name in(select process_name from perfil_process where user_id = 1) and cast(create_time as time) between cast('09:00:00' as time) and cast('19:00:00' as time) and status = 'running' group by process_name,process_id order by cpu_time desc;",conn)
process_time = process_time_detail.groupby(['process_name']).sum()

hi_webtime = int(process_time.cpu_time[(process_time.cpu_time.keys=='chrome')])
md_webtime = int(hi_webtime * 0.50)
lw_webtime = int(hi_webtime * 0.20)


util_worktime = int(process_time.cpu_time.sum())

total_domains = int(pd.read_sql_query("select count(domain) as qtd_domain from sites;",conn)["qtd_domain"])
valid_domains = int(pd.read_sql_query("select count(domain) as qtd_domain from sites where domain in (select domain from perfil_sites where user_id = 1);",conn)["qtd_domain"])
pct_sites = valid_domains / total_domains
pct_sites_util = int(hi_webtime * pct_sites)
```

# Fuzzy

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x_process = np.arange(0, hi_processtime, 1)
x_webtime = np.arange(0, hi_webtime, 1)
x_worktime  = np.arange(0, hi_worktime, 1)

process_lo = fuzz.trimf(x_process, [lw_processtime, lw_processtime, md_processtime])
process_md = fuzz.trimf(x_process, [lw_processtime, md_processtime, hi_processtime])
process_hi = fuzz.trimf(x_process, [md_processtime, hi_processtime, hi_processtime])

webtime_lo = fuzz.trimf(x_webtime, [lw_webtime, lw_webtime, md_webtime])
webtime_md = fuzz.trimf(x_webtime, [lw_webtime, md_webtime, hi_webtime])
webtime_hi = fuzz.trimf(x_webtime, [md_webtime, hi_webtime, hi_webtime])

worktime_lo = fuzz.trimf(x_worktime, [lw_worktime, lw_worktime, md_worktime])
worktime_md = fuzz.trimf(x_worktime, [lw_worktime, md_worktime, hi_worktime])
worktime_hi = fuzz.trimf(x_worktime, [md_worktime, hi_worktime, hi_worktime])

fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_process, process_lo, 'b', linewidth=1.5, label='Low')
ax0.plot(x_process, process_md, 'g', linewidth=1.5, label='Medium')
ax0.plot(x_process, process_hi, 'r', linewidth=1.5, label='High')
ax0.set_title('Process amount')
ax0.legend()

ax1.plot(x_webtime, webtime_lo, 'b', linewidth=1.5, label='Low')
ax1.plot(x_webtime, webtime_md, 'g', linewidth=1.5, label='Medium')
ax1.plot(x_webtime, webtime_hi, 'r', linewidth=1.5, label='High')
ax1.set_title('Webtime amount')
ax1.legend()

ax2.plot(x_worktime, worktime_lo, 'b', linewidth=1.5, label='Low')
ax2.plot(x_worktime, worktime_md, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_worktime, worktime_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('Worktime amount')
ax2.legend()

for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
process_level_lo = fuzz.interp_membership(x_process, process_lo, util_worktime)
process_level_md = fuzz.interp_membership(x_process, process_md, util_worktime)
process_level_hi = fuzz.interp_membership(x_process, process_hi, util_worktime)

webtime_level_lo = fuzz.interp_membership(x_webtime, webtime_lo, pct_sites_util)
webtime_level_md = fuzz.interp_membership(x_webtime, webtime_md, pct_sites_util)
webtime_level_hi = fuzz.interp_membership(x_webtime, webtime_hi, pct_sites_util)

active_rule1 = np.fmax(process_level_lo, webtime_level_lo)

worktime_activation_lo = np.fmin(active_rule1, worktime_lo)  # removed entirely to 0
worktime_activation_md = np.fmin(webtime_level_md, worktime_md)
active_rule3 = np.fmax(process_level_hi, webtime_level_hi)
worktime_activation_hi = np.fmin(active_rule3, worktime_hi)
worktime0 = np.zeros_like(x_worktime)


fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(x_worktime, worktime0, worktime_activation_lo, facecolor='b', alpha=0.7)
ax0.plot(x_worktime, worktime_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.fill_between(x_worktime, worktime0, worktime_activation_md, facecolor='g', alpha=0.7)
ax0.plot(x_worktime, worktime_md, 'g', linewidth=0.5, linestyle='--')
ax0.fill_between(x_worktime, worktime0, worktime_activation_hi, facecolor='r', alpha=0.7)
ax0.plot(x_worktime, worktime_hi, 'r', linewidth=0.5, linestyle='--')
ax0.set_title('Output worktime activity')

for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
aggregated = np.fmax(worktime_activation_lo,
                     np.fmax(worktime_activation_md, worktime_activation_hi))

worktime = fuzz.defuzz(x_worktime, aggregated, 'centroid')
worktime_activation = fuzz.interp_membership(x_worktime, aggregated, worktime)  # for plot


fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(x_worktime, worktime_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.plot(x_worktime, worktime_md, 'g', linewidth=0.5, linestyle='--')
ax0.plot(x_worktime, worktime_hi, 'r', linewidth=0.5, linestyle='--')
ax0.fill_between(x_worktime, worktime0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([worktime, worktime], [0, worktime_activation], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Aggregated worktime and result (line)')


for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```
